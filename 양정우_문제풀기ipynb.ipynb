<font size=5>회전 초밥집에서는 담긴 접시의 색에 따라 초밥의 가격이 다르게 계산된다.<br>다음을 참고하여 회전 초밥의 비용을 영수증 형식으로 출력하는 프로그램을 완성해 보자.<br>

<br><br><br>

##입력예시


```
빨간색 접시의 개수를 입력하세요: 2
노란색 접시의 개수를 입력하세요: 3
파란색 접시의 개수를 입력하세요: 1
초록색 접시의 개수를 입력하세요: 2
```


##출력예시
```
========= 회전 초밥 영수증 =======
 접시     단가     개수       합계
빨간색   1,000       2       2,000
노란색   2,000       3       6,000
파란색   3,000       1       3,000
초록색   4,000       2       8,000
총 가격                     19,000
==================================
```



입력=4개


변수=n개

input() print() 이스케이프시퀸스 계산
(합계)

````python
a=int(input("빨간색 접시의 개수는:"))
b=int(input("노란색 접시의 개수는:"))
c=int(input("파란색 접시의 개수는:"))
d=int(input("초록색 접시의 개수는:"))
print(a * 1000)
print(b * 2000)
print(c * 3000)
print(d * 4000)
````

````python
a=int(input("빨간색 접시의 개수를 입력하세요: "))
a_price = 1000
a_price_total = a * a_price
sushi_price_total = a_price_total

b=int(input("노란색 접시의 개수를 입력하세요: "))
b_price = 2000
b_price_total = b * b_price
sushi_price_total += b_price_total # 총합에 더하도록 수정

c=int(input("파란색 접시의 개수를 입력하세요: "))
c_price = 3000
c_price_total = c * c_price
sushi_price_total += c_price_total # 총합에 더하도록 수정

d=int(input("초록색 접시의 개수를 입력하세요: "))
d_price = 4000
d_price_total = d * d_price
sushi_price_total += d_price_total # 총합에 더하도록 수정

print("========= 회전 초밥 영수증 =======")
print(" 접시     단가     개수       합계")
print(f"빨간색   {a_price:,}       {a}       {a_price_total:,}")
print(f"노란색   {b_price:,}       {b}       {b_price_total:,}")
print(f"파란색   {c_price:,}       {c}       {c_price_total:,}")
print(f"초록색   {d_price:,}       {d}       {d_price_total:,}")
print(f"총 가격                     {sushi_price_total:,}")
print("==================================")
````
