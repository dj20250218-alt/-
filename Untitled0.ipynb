#영수증 제작

```python
a=int(input("몇개를 드셨습니까? (빨간색 접시를): "))
a_price = 10000
a_price_total = a * a_price
sushi_price_total=0
sushi_price_total=a_price_total
b=int(input("몇개를 드셨습니까? (노란색 접시를): "))
b_price = 20000
b_price_total = b * b_price
sushi_price_total += b_price_total # 총합에 더하도록 수정

c=int(input("파란색 접시의 개수를 입력하세요: "))
c_price = 30000
c_price_total = c * c_price
sushi_price_total += c_price_total # 총합에 더하도록 수정

d=int(input("초록색 접시의 개수를 입력하세요: "))
d_price = 4000
d_price_total = d * d_price
sushi_price_total += d_price_total # 총합에 더하도록 수정

print("========= 회전 초밥 영수증 =======")
print(" 접시     단가     개수       합계")
print(f"빨간색   {a_price:,}       {a}       {a_price_total:,}")
print(f"노란색   {b_price:,}       {b}       {b_price_total:,}")
print(f"파란색   {c_price:,}       {c}       {c_price_total:,}")
print(f"초록색   {d_price:,}       {d}       {d_price_total:,}")
print(f"총 가격                     {sushi_price_total:,}")
print("==================================")
```
