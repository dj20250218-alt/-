# 입력함수
input(): 키보드를 통해 입력반응을 보는 함수



```python
a=input("이름:")
b=input("나이:")
print(a + "님 안녕하세요")
print(20 - int(b),"년 후면 20살이시네요")
```

```python
a = input("가격(원):")
b = input("할인율(%):")

# Convert input strings to integers for calculation
price = int(a)
discount_rate = int(b)

# Calculate the discounted price
discounted_price = price * (1 - discount_rate / 100)

print(f"할인받은 가격이{discounted_price:0.0f}원 입니다.")
```

```python
#섭씨(℃)온도를 화씨온도 (F)로 바꾸는 프로그램

a= float(input("섭씨온도를 입력하세요"))

b=(9 * a / 5 )+ 32

print(f"섭씨{a}도는 화시 {b}도 입니다")
```
